```python
import matplotlib.pyplot as plt
import pandas as pd

# ----------------------------
# Models (same order as tables)
# ----------------------------
models = [
    'Random Forest', 'Gradient Boosting', 'SVM', 'KNN',
    'Decision Tree', 'LightGBM', 'Voting Classifier', 'MLP'
]

# ----------------------------
# Table 2: Without Class Balancing and Hyperparameter Tuning
# ----------------------------
accuracy_wo  = [0.9235, 0.9189, 0.9210, 0.9166, 0.8847, 0.8496, 0.9212, 0.9217]
f1_wo        = [0.9024, 0.9029, 0.8935, 0.9004, 0.8858, 0.8610, 0.8942, 0.9007]

# ----------------------------
# Table 3: With Class Balancing Using Random Oversampling
# ----------------------------
accuracy_os  = [0.9974, 0.9581, 0.9046, 0.9867, 0.9936, 0.9932, 0.9274, 0.9516]
f1_os        = [0.9974, 0.9574, 0.9019, 0.9865, 0.9935, 0.9932, 0.9257, 0.9509]

# DataFrame
df = pd.DataFrame({
    "Model": models,
    "Accuracy (No Balancing)": accuracy_wo,
    "Accuracy (Oversampling)": accuracy_os,
    "F1 (No Balancing)": f1_wo,
    "F1 (Oversampling)": f1_os
})

# Reverse order so MLP appears at the top
df_plot = df.iloc[::-1].reset_index(drop=True)

# ----------------------------
# Plot (2 rows x 2 columns) at 600 DPI
# ----------------------------
fig, ax = plt.subplots(2, 2, figsize=(14, 8), sharex=True, dpi=600)

# Accuracy plots
ax[0, 0].barh(df_plot["Model"], df_plot["Accuracy (No Balancing)"], color="lightblue")
ax[0, 0].set_title("Accuracy - Without Class Balancing")

ax[0, 1].barh(df_plot["Model"], df_plot["Accuracy (Oversampling)"], color="lightgreen")
ax[0, 1].set_title("Accuracy - With Class Balancing (Random Oversampling)")

# F1 plots
ax[1, 0].barh(df_plot["Model"], df_plot["F1 (No Balancing)"], color="lightblue")
ax[1, 0].set_title("F1 Score - Without Class Balancing")

ax[1, 1].barh(df_plot["Model"], df_plot["F1 (Oversampling)"], color="lightgreen")
ax[1, 1].set_title("F1 Score - With Class Balancing (Random Oversampling)")

# Formatting
for i in range(2):
    for j in range(2):
        ax[i, j].set_xlabel("Score")
        ax[i, j].set_ylabel("Models")
        ax[i, j].set_xlim(0, 1.0)

plt.tight_layout()

# Save at 600 DPI (recommended for papers)
plt.savefig("model_performance_600dpi.png", dpi=600, bbox_inches="tight")

plt.show()
```
